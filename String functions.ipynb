## Lower - it converts strings in the Series to lower case

```python
Ecom["Lower_City"]= Ecom.City.str.lower()
```

```python
Ecom[['City','Lower_City']].tail()
```

```python
pd.set_option("display.max_columns",100)
Ecom.head(3)
```

## Upper- it converts strings in the series to upper case.

```python
Ecom["Cu_Na"] = Ecom["Customer Name"].str.upper()
```

```python
Ecom[["Customer Name","Cu_Na"]].head()
```

```python
pd.set_option("display.max_columns",100)
Ecom.head()
```

## capitalize - It converts first character to uppercase and remaining to lowercase

```python
Ecom["Cu_Na"].head()
```

```python
Ecom["Customer_Capitalize"] = Ecom["Cu_Na"].str.capitalize()
```

```python
Ecom[["Customer Name","Cu_Na","Customer_Capitalize"]].head()
```

```python
pd.set_option("display.max_columns",100)
Ecom.head(3)
```

## casefold - Removes all case distinctions in the string.

```python
Ecom["Casefold"] = Ecom["City"].str.casefold()
```

```python
Ecom[["City","Casefold"]].head()
```

```python
pd.set_option("display.max_columns",100)
Ecom.head(2)
```

## Length - It gives the  string length

```python
Ecom["Length_Name"] = Ecom["Customer Name"].str.len()
```

```python
Ecom[["Customer Name","Length_Name"]].tail()
```

##  split- It splits the string in the Series

```python
Ecom["split"] = Ecom["Order ID"].str.split('-')
```

```python
Ecom["split"].head()
```

```python
pd.set_option("display.max_columns",100)
Ecom.head()
```

## Startswith

## Find - It is used to search a substring in each string present in a series (position of value-1)

```python
Ecom["Find"]= Ecom["Country"].str.find('i')
```

```python
Ecom[["Country","Find"]].head()
```

## swapcase-lower case characters will be converted into uppercase and uppercase character into lower case in every string.

```python
Ecom["Swap"] = Ecom["Market"].str.swapcase()
```

```python
Ecom[["Market","Swap"]].head(10)
```

## islower-If all characters in the string are lowercase

```python
Ecom["islower"] = Ecom["Segment"].str.islower()
```

```python
Ecom[["Segment","islower"]].head(10)
```

## isupper-If all characters in the string are uppercase

```python
Ecom["isupper"] = Ecom["Market"].str.isupper()
```

```python
Ecom[["isupper" ,"Market"]].head(10)
```

## isnumeric -  It returns true if all characters in a string are numeric characters. If not, it returns False

```python
Ecom["isnumeric"] = Ecom["Customer ID"].str.isnumeric()
```

```python
Ecom[["Customer ID","isnumeric"]].head()
```

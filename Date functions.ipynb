###exctract only day short version

```python
Ecom["Order Day"] = Ecom["Order Date"].apply(lambda x: x.strftime('%a'))
```

```python
Ecom.head()
```

### exctract only day full version

```python
Ecom["Order Day2"] = Ecom["Order Date"].apply(lambda x: x.strftime('%A'))
```

```python
Ecom["Order Day2"]
```

## Weekday as a number 0-6, 0 is Sunday

```python
Ecom["Order week"] = Ecom["Order Date"].apply(lambda x: x.strftime('%w'))
```

```python
Ecom[["Order Date","Order week"]].head()
```

## Day of month 01-31

```python
Ecom["Order month"] = Ecom["Order Date"].apply(lambda x: x.strftime('%d'))
```

```python
Ecom[["Order Date","Order month"]].head()
```

## Month name, short version

```python
Ecom["Order month1"] = Ecom["Ship Date"].apply(lambda x: x.strftime('%b'))
Ecom[["Order Date","Order month1"]].head()
```

## Month name, Full version

```python
Ecom["Order month2"] = Ecom["Ship Date"].apply(lambda x: x.strftime('%B'))
Ecom[["Order Date","Order month2"]].head()
```

## Month as a number 01-12

```python
Ecom["Order month3"] = Ecom["Ship Date"].apply(lambda x: x.strftime('%m'))
Ecom[["Order Date","Order month3"]].head()
```

## Year- short version, without century

```python
Ecom["Order Year"] = Ecom["Order Date"].apply(lambda x: x.strftime('%y'))
Ecom[["Order Date","Order Year"]].head()
```

## Year-Full version

```python
Ecom["Order Year2"] = Ecom["Ship Date"].apply(lambda x: x.strftime('%Y'))
Ecom[["Order Date","Order Year2"]].head()
```

## Day number of year 001-366

```python
Ecom["Order Day3"] = Ecom["Order Date"].apply(lambda x: x.strftime('%j'))
Ecom[["Order Date","Order Day3"]].head(10)
```

## Week number of year, Sunday as the first day of week, 00-53

```python
Ecom["Order Week"] = Ecom["Order Date"].apply(lambda x: x.strftime('%U'))
Ecom[["Order Date","Order Week"]].head(10)
```

## Week number of year, Monday as the first day of week, 00-53

```python
Ecom["Order Week2"] = Ecom["Order Date"].apply(lambda x: x.strftime('%W'))
Ecom[["Order Date","Order Week2"]].head(10)
```

## Local version of date and time

```python
Ecom["Order date and time"] = Ecom["Order Date"].apply(lambda x: x.strftime('%c'))
Ecom[["Order Date","Order date and time"]].head(10)
```

## Century

```python
Ecom["Order Century"] = Ecom["Order Date"].apply(lambda x: x.strftime('%C'))
Ecom[["Order Date","Order Century"]].head(10)
```

## Local version of date

```python
Ecom["Order Date2"] = Ecom["Order Date"].apply(lambda x: x.strftime('%x'))
Ecom[["Order Date","Order Date2"]].head(10)
```
